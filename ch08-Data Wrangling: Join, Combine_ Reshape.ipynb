## Combining and Merging Datasets(pd.merge, pd.concat, combine_first)
* pd.merge: Merge or join operations combine datasets by linking rows using one or more keys.
* pd.concat: “stacks” together objects along an axis.
* combine_first: splicing together overlapping data to fill in missing values in one object with values from another.

### Database-Style DataFrame Joins
* pd.merge(df1, df2, on='key')
* pd.merge(df3, df4, left_on='lkey', right_on='rkey')
* pd.merge(df1, df2, how='outer') 
* pd.merge(df1, df2, on='key', how='left')
* pd.merge(df1, df2, how='inner')

### many to one merge(join)

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df1
```

```python
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
df2
```

```python
#many-to-one join
pd.merge(df1, df2) #common key only merged.
```

### good practice to specify explicitly which column to join on:
Even though information is not specified, merge uses the overlapping column names as the keys.
* pd.merge(df1, df2, on='key')
* pd.merge(df3, df4, left_on='lkey', right_on='rkey')
* pd.merge(df1, df2, how='inner') #intersection #pd.merge(df1, df2)
* pd.merge(df1, df2, how='outer') #union 
* pd.merge(df1, df2, on='key', how='left')


```python
pd.merge(df1, df2, on='key')
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df3
```

```python
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
df4
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey') #common key only (intersection)
```

```python
pd.merge(df1, df2, how='outer') #outer : all key either df1 or df2 (union)
```

### many to many merge

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                    'data1': range(6)})
df1
```

### Merging on Index (right1, index=['a', 'b'])

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                      'value': range(6)})  #column 'key' with no specified index
left1
```

```python
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b']) # with index
right1
```

* merge fnction arguments: 
    * left_on : Columns in left DataFrame to use as join keys.
    * left_index : Use row index in left as its join key (or keys, if a MultiIndex).
* the merge key(s) in a DataFrame will be found in its index.(right1 -> a,b) 
    * In this case, you can pass left_index=True or right_index=True (or both) to indicate that
the index should be used as the merge key:

```python
pd.merge(left1, right1, left_on='key', right_index=True) #left1 df's key is on
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer') #uinon
```

####  join also supports joining the index of the passed DataFrame(right1's index- a,b ) on one of the columns of the calling DataFrame(left1's column, 'key' - a,a,a,b,b)
**join** :  convenient merging by index. (left join on the join keys :  exactly preserving the left frame’s row
index) 
* combine together many DataFrame objects having the same or similar indexes but non-overlapping columns.

### Pivoting “Wide” to “Long” Format : pd.melt <-> df.pivot
* pd.melt : merges multiple columns into one, producing a DataFrame that is longer than the input.
* df.pivot : transforming one column into many in a new DataFrame

```python
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],
                   'A': [1, 2, 3],
                   'B': [4, 5, 6],
                   'C': [7, 8, 9]})
df
```

```python
melted = pd.melt(df, ['key'])
melted
```

```python
#trasform variable, column to distinct values(A,B,C) indexed by key column.
reshaped = melted.pivot('key', 'variable', 'value')
reshaped
```

**reset_index** to move the data back into a column
since pivot creates an index(key) from the column used as the row labels.

```python
reshaped.reset_index() 
```

specify a subset of columns to use as value columns:

```python
pd.melt(df, id_vars=['key'], value_vars=['A', 'B'])
```

pandas.melt can be used without any group identifiers, too:

```python
pd.melt(df, value_vars=['A', 'B', 'C'])
```
